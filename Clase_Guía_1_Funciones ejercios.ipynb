```python
# importamos pyplot de matplotlib y numpy
import matplotlib.pyplot as plt
import numpy as np
```

Se genera una tabla de datos usando numpy a la que
se aplicar√° la funci√≥n dada en el ejemplo, se toma la variable independiente x entre 0 y 6
y los elementos de la tabla se separaran en 0,01.

Para cada elemento x, se aplic√≥ la funci√≥n. Luego se arreglan los coeficientes de los ejes, se agrega un t√≠tulo al gr√°fico, t√≠tulos de los ejes y una grilla.

```python
#Variable independiente x entre 0 y 6.
x = np.arange(0, 6, 0.01)
# para cada elemento, se aplicar√° la funci√≥n
y = 120*x+2000
# le pedimos a matplotlib que genere un gr√°fico
plt.plot(x, y)
# cuyos ejes van de 0 a 6 en la variable x, de 0 a 2800 en la variable y
plt.axis([0, 6, 0, 2800])
# con un t√≠tulo de eje x
plt.xlabel('Energ√≠a utilizada ({kWh})')
# t√≠tulo de eje y
plt.ylabel('Costo para usuario ($)')
# titulo del gr√°fico
plt.title('Costo para el usuario seg√∫n uso de energ√≠a')
# se pide mostrar una grilla para facilitar la lectura
plt.grid(True)
# Finalmente, que muestre el gr√°fico
plt.show()
```
